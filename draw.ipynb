```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("time_log",'r') as file:
    lines = file.readlines()
    n_list = set()
    k_list = set()  
    time_list = []
    for line in lines:
        tmp = line.split('  ')
        n_list.add(int(tmp[0]))
        k_list.add(int(tmp[1]))
        time_list.append(float(tmp[2]))
```

```python
X=np.sort(np.array(list(n_list)))
Y=np.sort(np.array(list(k_list)))
print(X)
print(Y)
X, Y = np.meshgrid(X, Y)
Z=np.array(time_list).reshape(3,5).T
Z = Z*1000
print(Z)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='cividis')
ax.set_xlabel('n')
ax.set_ylabel('k')
ax.set_zlabel('time(ms)')
plt.tight_layout()
```
